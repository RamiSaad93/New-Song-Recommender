```python
import pandas as pd
import random
import numpy as np

#Visualisation
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Web Scraping
from bs4 import BeautifulSoup
import requests

# Data Processing
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Unsupervised ML
from sklearn.cluster import KMeans
from matplotlib.lines import Line2D
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
```

# 1- WEB SCRAPING

```python
url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

### (A) Artist Name

```python
artists_names = []
artists_all = soup.find_all('span', class_ = "a-children-link-color-black")
for i in artists_all:
    artists_names.append(i.get_text().strip())

artists_names
```

### (b) Song Name

```python
song_names_all = []
song_name = soup.find_all('h3', class_ = "u-max-width-397")
for i in song_name:
    song_names_all.append(i.get_text().strip())
song_names_all
```

### (c) Song Number

```python
song_number = []
number_all = soup.find_all('span', class_ = "u-line-height-33px")
for i in number_all:
    song_number.append(i.get_text().strip())

song_number
```

```python
billboard_df = pd.DataFrame({"Number": song_number, "Song": song_names_all, "Artist": artists_names})
billboard_df.head(60)
```

```python
billboard_df.to_csv("billboard_df.csv")
```

```python
# Scaling, outlier handling, and feature selection/dimensionality reduction before clustering
# KMeans benefits more from StandardScaler, DBSCAN often from MinMaxScaler
```

```python
features = ['popularity', 'duration_ms', 'danceability', 'energy', 'key',
            'loudness', 'mode', 'speechiness', 'acousticness',
            'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']

scaler = StandardScaler()
X_std_scaled_arrays = scaler.fit_transform(af_df[features])
X_std_scaled = pd.DataFrame(X_std_scaled_arrays, columns = features)
X_std_scaled = pd.concat([af_df[['track_id', 'artists', 'album_name', 'track_name']], X_std_scaled], axis = 1)
X_std_scaled
```

```python
z_scores = np.abs(X_std_scaled_arrays)
threshold = 3
filter_ = (z_scores < threshold).all(axis=1)
```

```python
X_no_outliers = X_std_scaled_arrays[filter_]
af_df_no_outliers = af_df[filter_] 
```

```python
af_df_no_outliers.reset_index(drop=True, inplace=True)
```

```python
af_df_no_outliers.shape
```

```python
string_features = af_df_no_outliers.select_dtypes(include = 'object')
numerical_features = af_df_no_outliers.select_dtypes(include = 'number')
```

```python
string_features
```

```python
numerical_features
```

### (B) Scaling

#### - MinMaxScaler

```python
normalizer = MinMaxScaler()
normalizer.fit(numerical_features)
X_train_norm = pd.DataFrame(normalizer.transform(numerical_features), columns = numerical_features.columns)
X_train_norm
```

#### - StandardScaler

```python
std_scaler = StandardScaler()
std_scaler.fit(numerical_features)
X_train_scaled = pd.DataFrame(std_scaler.transform(numerical_features), columns = numerical_features.columns)
X_train_scaled
```

### (C) Starting The Model

#### - Normalized Data

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20))
visualizer.fit(X_train_norm)
visualizer.poof()
```

```python
cluster_range = range(4,25)
sil_score_list_my = []
for i in cluster_range:
    
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(X_train_norm)

    sil_score = silhouette_score(X_train_norm, labels)
    sil_score_list_my.append(sil_score)
    
xpoints = np.array(cluster_range)
ypoints = np.array(sil_score_list_my)

plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

#### - Scaled Data

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20))
visualizer.fit(X_train_scaled)
visualizer.poof()
```

```python
cluster_range = range(5,20)
sil_score_list_my = []
for i in cluster_range:
    
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(X_train_scaled)

    sil_score = silhouette_score(X_train_scaled, labels)
    sil_score_list_my.append(sil_score)
    
xpoints = np.array(cluster_range)
ypoints = np.array(sil_score_list_my)

plt.plot(xpoints, ypoints)
plt.show()
```

#### - Creating the Labels

```python
kmeans = KMeans(n_clusters=6, random_state=42)
labels = kmeans.fit_predict(X_train_norm)
sil_score = silhouette_score(X_train_norm, labels)
print(sil_score)
numerical_features['Labels'] = labels
numerical_features
```

```python
numerical_features['Labels'].value_counts()
```

```python
Whole_df = pd.concat([string_features , numerical_features], axis = 1)
Whole_df
```

# PROGRAM

```python
def song_recommender():
    refrence = input("Are you interested in a Trending Now song or a specific Genre?").lower()
    if refrence == "trending now":
        recommended_song = random.choice(billboard_df['Song'])
        artist_of_recommended_song = billboard_df.loc[billboard_df["Song"] == recommended_song, 'Artist'].values[0]
        number_of_song = billboard_df.loc[billboard_df["Song"] == recommended_song, 'Number'].values[0]
        print(f"We can recommend you this song: ({recommended_song}) by ({artist_of_recommended_song})")
        print(f"It is very trending right now, as it is actually number {number_of_song} on the list of the top 100 Billboard song")
        recommended_billboard_song_id = sp.search(q=recommended_song, type='track',market="GB")["tracks"]["items"][0]["id"]
        return play_song(recommended_billboard_song_id)
    else:
        return print("We still working on it")
```
